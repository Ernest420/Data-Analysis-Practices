```python
# Load the dataset from Unity Catalog Volume
catalog = "workspace"
schema = "info-diabetes"
volume = "datos-diabetes"
file_name = "diabetes.csv"

file_path = f"/Volumes/{catalog}/{schema}/{volume}/{file_name}"

df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .option("delimiter", ",")
    .csv(file_path)
)

# Información general del dataset
print("=== INFORMACIÓN GENERAL ===")
df.printSchema()

# Mostrar los nombres de las columnas para verificar
print("Nombres de columnas:", df.columns)

# Estadísticas descriptivas
print("=== ESTADÍSTICAS DESCRIPTIVAS ===")
display(df.describe())

# Verificar valores nulos o cero en columnas clave
print("=== VALORES CERO EN COLUMNAS CLAVE ===")
from pyspark.sql.functions import col

# Vamos a usar las columnas que existen en el DataFrame. Si no existe 'Outcome', no la incluiremos en columns_to_check.
# Pero primero, definamos columns_to_check con las columnas numéricas que son comunes en el dataset de diabetes.
# Nota: El dataset de diabetes típico tiene: Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age, Outcome.
# Pero si no está Outcome, entonces no la usamos.

columns_to_check = [
    'Glucose',
    'BloodPressure',
    'SkinThickness',
    'Insulin',
    'BMI'
]

# Filtrar solo las columnas que existen en el DataFrame
existing_columns_to_check = [col for col in columns_to_check if col in df.columns]

for column in existing_columns_to_check:
    zero_count = df.filter(col(column) == 0).count()
    print(f"{column}: {zero_count} valores cero")

# Distribución de la variable objetivo
print("=== DISTRIBUCIÓN DE LA VARIABLE OBJETIVO (Outcome) ===")
if "Outcome" in df.columns:
    total_count = df.count()
    outcome_dist = (
        df.groupBy("Outcome")
        .count()
        .withColumn("Porcentaje", (col("count") / total_count) * 100)
    )
    display(outcome_dist)
else:
    print("La columna 'Outcome' no se encuentra en el dataset. Las columnas disponibles son:", df.columns)
```

```python
from pyspark.sql.functions import when

# Calcular medianas para las columnas con valores cero
medians = {}
for column in columns_to_check:
    median_val = df.approxQuantile(column, [0.5], 0.01)[0]
    medians[column] = median_val
    print(f"Mediana de {column}: {median_val}")

# Imputar valores cero con la mediana
df_clean = df
for column, median_val in medians.items():
    df_clean = df_clean.withColumn(column, when(col(column) == 0, median_val).otherwise(col(column)))

# Verificar que no quedan valores cero en estas columnas
print("=== VERIFICACIÓN POST-IMPUTACIÓN ===")
for column in columns_to_check:
    zero_count = df_clean.filter(col(column) == 0).count()
    print(f"{column}: {zero_count} valores cero")
```

```python
# Convertir a Pandas para visualizaciones (si el dataset no es muy grande)
pdf = df_clean.toPandas()

# Importar librerías para visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Intentar establecer un estilo por defecto, si falla, omitirlo
try:
    plt.style.use('default')
except:
    pass

# Verificar las columnas disponibles en pdf
print("Columnas en el DataFrame pdf:", pdf.columns.tolist())

# Si no existe 'Outcome', buscar una alternativa
outcome_column = None
if 'Outcome' in pdf.columns:
    outcome_column = 'Outcome'
else:
    # Buscar alguna columna que pueda ser la variable objetivo
    possible_names = ['Outcome', 'Class', 'Diabetes', 'Target', 'Result']
    for name in possible_names:
        if name in pdf.columns:
            outcome_column = name
            break

# Si no se encuentra, usar la última columna como variable objetivo (común en datasets)
if outcome_column is None:
    outcome_column = pdf.columns[-1]
    print(f"Usando la última columna como variable objetivo: {outcome_column}")

# Histogramas para variables numéricas
numeric_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
# Asegurarse de que las columnas numéricas existen en pdf
numeric_cols = [col for col in numeric_cols if col in pdf.columns]

pdf[numeric_cols].hist(figsize=(15, 10), bins=20)
plt.suptitle('Distribución de Variables Numéricas')
plt.show()

# Matriz de correlación
plt.figure(figsize=(12, 8))
corr_matrix = pdf.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()

# Si tenemos una columna para el outcome, hacer boxplots por esa variable
if outcome_column is not None:
    fig, axes = plt.subplots(2, 4, figsize=(20, 10))
    for i, col in enumerate(numeric_cols):
        row, col_idx = i // 4, i % 4
        sns.boxplot(data=pdf, x=outcome_column, y=col, ax=axes[row, col_idx])
        axes[row, col_idx].set_title(f'Boxplot de {col} por {outcome_column}')
    plt.tight_layout()
    plt.show()
else:
    print("No se encontró una columna para la variable objetivo (Outcome) para los boxplots.")
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator

print("=== INICIANDO MODELADO CON COLUMNA 'CLASE' ===")

# Verificar las columnas reales
print("Columnas en df_clean:", df_clean.columns)

# PREPARAR FEATURES - usar "Clase" en lugar de "Outcome"
feature_columns = [col for col in df_clean.columns if col != "Clase"]
print(f"Variables predictoras: {feature_columns}")
print(f"Variable objetivo: Clase")

# Crear el assembler
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
feature_data = assembler.transform(df_clean)

print("✅ Datos transformados con features")
display(feature_data.select("Clase", "features").limit(5))

# DIVIDIR EN ENTRENAMIENTO Y PRUEBA
train_data, test_data = feature_data.randomSplit([0.8, 0.2], seed=42)

print(f"📊 Datos de entrenamiento: {train_data.count()} registros")
print(f"📊 Datos de prueba: {test_data.count()} registros")

# ENTRENAR EL MODELO - usar labelCol="Clase"
lr = LogisticRegression(featuresCol="features", labelCol="Clase", maxIter=10)
print("🔄 Entrenando modelo de regresión logística...")
lr_model = lr.fit(train_data)

print("✅ Modelo entrenado exitosamente")

# EVALUAR EL MODELO - usar labelCol="Clase"
predictions = lr_model.transform(test_data)

evaluator = BinaryClassificationEvaluator(
    labelCol="Clase", 
    rawPredictionCol="rawPrediction", 
    metricName="areaUnderROC"
)

auc = evaluator.evaluate(predictions)
print(f"🎯 Área bajo la curva ROC: {auc:.3f}")

# Mostrar algunas predicciones
print("\n🔍 Primeras predicciones:")
display(predictions.select("Clase", "prediction", "probability").limit(10))

# Matriz de confusión
print("\n📊 Matriz de confusión:")
conf_matrix = predictions.groupBy("Clase", "prediction").count()
display(conf_matrix)

```

```python
# IMPORTAR LAS LIBRERÍAS NECESARIAS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("=== ANÁLISIS DE IMPORTANCIA DE CARACTERÍSTICAS ===")

# Verificar que tenemos el modelo entrenado
try:
    # Obtener coeficientes del modelo
    coefficients = lr_model.coefficients.toArray()
    print("✅ Coeficientes obtenidos del modelo")
    
    # Crear DataFrame de importancia
    feature_importance = pd.DataFrame({
        'Característica': feature_columns,
        'Importancia': abs(coefficients),  # Valor absoluto para importancia
        'Coeficiente': coefficients        # Coeficiente original (dirección)
    }).sort_values('Importancia', ascending=False)
    
    print("\n📊 Ranking de importancia de características:")
    display(feature_importance)
    
    # VISUALIZACIÓN MEJORADA
    plt.figure(figsize=(12, 8))
    
    # Crear subplots
    plt.subplot(1, 2, 1)
    sns.barplot(data=feature_importance, x='Importancia', y='Característica', palette='viridis')
    plt.title('Importancia Absoluta de Características\n(Regresión Logística)')
    plt.xlabel('Importancia (valor absoluto del coeficiente)')
    
    plt.subplot(1, 2, 2)
    # Gráfico de coeficientes con colores por signo
    colors = ['red' if x < 0 else 'green' for x in feature_importance['Coeficiente']]
    sns.barplot(data=feature_importance, x='Coeficiente', y='Característica', palette=colors)
    plt.title('Coeficientes Originales\n(Rojo: negativo, Verde: positivo)')
    plt.xlabel('Coeficiente (dirección del efecto)')
    
    plt.tight_layout()
    plt.show()
    
    # INTERPRETACIÓN CLÍNICA
    print("\n🎯 INTERPRETACIÓN CLÍNICA:")
    print("=" * 50)
    
    top_features = feature_importance.head(3)
    for _, row in top_features.iterrows():
        feature = row['Característica']
        coef = row['Coeficiente']
        importance = row['Importancia']
        
        if coef > 0:
            effect = "AUMENTA el riesgo de diabetes"
        else:
            effect = "DISMINUYE el riesgo de diabetes"
            
        print(f"• {feature}: {effect} (importancia: {importance:.4f})")
    
    # Explicación de las variables más importantes
    print("\n📖 EXPLICACIÓN DE VARIABLES CLAVE:")
    feature_explanations = {
        'Glucose': 'Niveles elevados de glucosa en sangre son el indicador más fuerte de diabetes',
        'BMI': 'Índice de Masa Corporal alto está asociado con resistencia a la insulina',
        'Age': 'La edad avanzada es un factor de riesgo conocido para diabetes tipo 2',
        'Pregnancies': 'Embarazos múltiples pueden aumentar el riesgo de diabetes gestacional',
        'DiabetesPedigreeFunction': 'Historial familiar de diabetes (factor genético)',
        'Insulin': 'Niveles de insulina relacionados con la función pancreática',
        'BloodPressure': 'Presión arterial alta asociada con síndrome metabólico',
        'SkinThickness': 'Medida de grasa corporal relacionada con resistencia a insulina'
    }
    
    for feature in feature_importance['Característica'].head(5):
        if feature in feature_explanations:
            print(f"  {feature}: {feature_explanations[feature]}")
    
except Exception as e:
    print(f"❌ Error en el análisis: {e}")
    print("\n🔧 SOLUCIÓN ALTERNATIVA:")
    
    # Solución alternativa básica
    if 'coefficients' in locals():
        print("Coeficientes disponibles, mostrando lista simple:")
        for feature, coef in zip(feature_columns, coefficients):
            print(f"  {feature}: {coef:.4f}")
    else:
        print("No se encontraron coeficientes. Verifica que el modelo se entrenó correctamente.")
```

```python
# ANÁLISIS COMPLETO DE PERFORMANCE - CÓDIGO CORREGIDO
print("=== ANÁLISIS DETALLADO DE PERFORMANCE ===")

from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.sql.functions import col  # Asegurar importación

# Reiniciar el evaluador por si hay conflicto de nombres
evaluator = MulticlassClassificationEvaluator(
    labelCol="Clase", 
    predictionCol="prediction"
)

# Definir métricas
metric_names = {
    "Exactitud (Accuracy)": "accuracy",
    "Precisión (Precision)": "weightedPrecision", 
    "Recall": "weightedRecall",
    "F1-Score": "f1"
}

print("\n📈 MÉTRICAS DE CLASIFICACIÓN:")

# Calcular cada métrica individualmente
for metric_name, metric_param in metric_names.items():
    try:
        # Configurar la métrica y evaluar
        evaluator.setMetricName(metric_param)
        value = evaluator.evaluate(predictions)
        print(f"  {metric_name}: {value:.3f}")
    except Exception as e:
        print(f"  {metric_name}: Error - {e}")

# ANÁLISIS POR CLASE - CÓDIGO CORREGIDO
print("\n🎯 ANÁLISIS POR CLASE:")
for class_label in [0, 1]:
    try:
        class_predictions = predictions.filter(col("Clase") == class_label)
        class_count = class_predictions.count()
        correct_predictions = class_predictions.filter(col("prediction") == class_label).count()
        
        if class_count > 0:
            accuracy_class = correct_predictions / class_count
            class_description = "Sin Diabetes" if class_label == 0 else "Con Diabetes"
            print(f"  {class_description} (Clase {class_label}): {accuracy_class:.1%} de accuracy ({correct_predictions}/{class_count})")
        else:
            print(f"  Clase {class_label}: No hay ejemplos en los datos de prueba")
    except Exception as e:
        print(f"  Error analizando clase {class_label}: {e}")
```
